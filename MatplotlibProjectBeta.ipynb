### Line graph

```python
gas= pd.read_csv("Desktop/gas_prices.csv")

#gas.head()
```

```python
#Resize graph
plt.figure(figsize=(8,5))


plt.plot(gas['Year'],gas['USA'],'r.-',label="USA")
plt.plot(gas['Year'],gas['Canada'],'b.-',label="Canada")
plt.plot(gas['Year'],gas['South Korea'],'c.-',label="South Korea")
plt.plot(gas['Year'],gas['Australia'],'g.-',label="Australia")

#X and Y Labels
plt.xlabel('Year',fontsize=12)
plt.ylabel('Price',fontsize=12)

#add legend
plt.legend()
#add title
plt.title('Gas Prices over Time',fontsize=15,fontname="Calibri")


plt.xticks(gas['Year'][::2].tolist()+[2011])
price=list(range(7))
plt.yticks(price, [f"{x} USD" for x in price])

#add grid
plt.grid(color='gray',linestyle='--',linewidth='0.5')

plt.savefig('Gas_price_figure.png',dpi=300)

#show graph
plt.show()
```

```python
#column_names = list(gas.columns.values)
countries = list(gas.columns.values)[1:]

meanprices=gas[countries].mean()
#print(meanprices)
#meandf=pd.DataFrame(meanprices)
#kolumny=list(meandf.columns.values)
#print(meanprices)



#meandata=meandf[meandf.columns[]]
#meandata.head()

plt.scatter(countries,meanprices)
plt.title("Mean Gas Price by Country")
plt.xticks(rotation='vertical')

numbers = np.arange(1.5, 5.5, 0.5)
plt.yticks(numbers,[f"{x} USD" for x in numbers])
plt.xlabel("Country")
plt.ylabel("Price")

for country, meanprice in zip(countries, meanprices):
    plt.annotate(f'{meanprice:.2f}', (country, meanprice), textcoords="offset points", xytext=(0,5), ha='center')

plt.figure(figsize=(7,5))

plt.show()
```
